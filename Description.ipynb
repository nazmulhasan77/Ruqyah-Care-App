## 📱 **Ruqyah Care App**

**ডেভেলপার:** Butterfly Devs

**ধরন:** ইসলামিক / রুকইয়াহ ভিত্তিক অ্যাপ 

**উদ্দেশ্য:** কুরআন ও সুন্নাহ অনুযায়ী **রুকইয়াহ (ইসলামিক চিকিৎসা/হিলিং)** শেখানো ও প্রয়োগে সাহায্য করা

---

## 🧠 অ্যাপটি কী কাজ করে?

এটা শুধু দোয়ার অ্যাপ না — এটা একটা **Complete Ruqyah Toolkit**

### 🔹 মূল ফিচারগুলো

✅ **মাসনুন দোয়া ও যিকির**

* দৈনন্দিন জীবনের প্রয়োজনীয় দোয়া
* সুরক্ষা (protection) এর জন্য দোয়া

✅ **রুকইয়াহ আয়াত (Qur’anic Verses)**

* জিন, বদ নজর, জাদু ইত্যাদির জন্য নির্দিষ্ট আয়াত

✅ **অডিও রিসাইটেশন**

* রুকইয়াহ ও কুরআন শুনতে পারবে

✅ **গাইডলাইন / নির্দেশনা**

* কিভাবে রুকইয়াহ করতে হয় (step-by-step)

✅ **Self Diagnosis (স্বয়ং যাচাই)**

* কিছু প্রশ্ন/লক্ষণ দেখে বুঝতে সাহায্য করে:

  * জিনের সমস্যা
  * জাদু (sihr)
  * বদ নজর

✅ **ভিডিও কনটেন্ট**

* রুকইয়াহ বিষয়ক ইসলামিক ভিডিও
